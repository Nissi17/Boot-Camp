```python
num = [132,545,434,66,44,33,212,8]

even_count = 0
odd_count = 0

for n in num:
    if n%2==0:
        even_count += 1
    else:
        odd_count += 1

print("Count of even :", even_count)
print("Count of odd : ", odd_count)
```
